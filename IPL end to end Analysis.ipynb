# Importing Libraries and Dataset

# Maximum Dot Balls Faced Team Per Year

```python
dot_balls=new[new['batsman_runs']==0]
```

```python
dot_balls.groupby(['season','batting_team'])['batsman_runs'].count().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')
```

```python
ABC=dot_balls.groupby(['over','batting_team'])['batsman_runs'].count().sort_values(ascending=False).reset_index().drop_duplicates(subset='over',keep='first').sort_values('over')
```

```python
ABC.rename(columns={'batsman_runs':'dot_balls'},inplace=True)
```

```python
ABC.set_index('over',inplace=True)
```

# Batsman Profile

```python
def run_scored(batsman_name):
    batsman=delivery[delivery['batsman']==batsman_name]
    return batsman.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
run_scored('V Kohli')
```

# Most Destructive Batsman at Death Overs

```python
death_over=delivery[delivery['over']>15]
all_batsmans=death_over.groupby('batsman')['batsman_runs'].count()
x=all_batsmans>200
final=all_batsmans[x].index.tolist()
death_balls=all_batsmans[x]
total=delivery[delivery['batsman'].isin(final)]
death_over_final_batsman=total[total['over']>15]
final_batsmans=death_over_final_batsman.groupby('batsman')['batsman_runs'].sum()
sr=(final_batsmans/death_balls)*100
sr.sort_values(ascending=False).head(10)
```

# Wickets Taken by Bowler

```python
def wickets_taken(bowler_name):
    C=delivery[delivery['bowler']==bowler_name]
    return C.groupby('bowler')['player_dismissed'].count().values[0]
```

```python
wickets_taken('R Ashwin')
```
